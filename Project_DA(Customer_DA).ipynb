# Cleaning

```python
cleaned_df = df.copy()
```

```python
cleaned_df["Invoice"] = cleaned_df["Invoice"].astype("str")

mask = (
    cleaned_df["Invoice"].str.match("^\\d{6}$") == True

)

cleaned_df = cleaned_df[mask]

cleaned_df
```

```python
cleaned_df["StockCode"] = cleaned_df["StockCode"].astype("str")

mask= (
     (cleaned_df["StockCode"].str.match('^\\d{5}$') == True)
    |(cleaned_df["StockCode"].str.match('^\\d{5}[a-zA-z]+$') == True)
    |(cleaned_df["StockCode"].str.match('^PADS$') == True)
)

cleaned_df = cleaned_df[mask]

cleaned_df
```

# Feature Engineering

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"]*cleaned_df["Price"]

cleaned_df
```

```python
agg_df = cleaned_df.groupby(by="Customer ID", as_index=False) \
      .agg(
          MonetaryValue = ("SalesLineTotal","sum"),
          Frequency=("Invoice","nunique"),
          LastInvoiceDate = ("InvoiceDate", "max")
      )

agg_df.head(5)
```

```python
max_invoice_date = agg_df["LastInvoiceDate"].max()

agg_df["Recency"] = (max_invoice_date - agg_df["LastInvoiceDate"]).dt.days

agg_df.head(5)
```

```python
m_Q1 = agg_df["MonetaryValue"].quantile(0.25)
m_Q3 = agg_df["MonetaryValue"].quantile(0.75)
m_IQR = m_Q3 - m_Q1

Monetary_outliers_df = agg_df[(agg_df["MonetaryValue"] > m_Q3 + 1.5 * m_IQR) | (agg_df["MonetaryValue"] < m_Q1 - 1.5 * m_IQR) ].copy()

Monetary_outliers_df.describe()

```

```python
m_Q1 = agg_df["Frequency"].quantile(0.25)
m_Q3 = agg_df["Frequency"].quantile(0.75)
m_IQR = m_Q3 - m_Q1

frequency_outliers_df = agg_df[(agg_df["Frequency"] > m_Q3 + 1.5 * m_IQR) | (agg_df["Frequency"] < m_Q1 - 1.5 * m_IQR) ].copy()

frequency_outliers_df.describe()
```

```python
non_outliers_df = agg_df[(~agg_df.index.isin(Monetary_outliers_df.index)) & (~agg_df.index.isin(frequency_outliers_df.index))]

non_outliers_df.describe()
```

# KMeans Clustring

# K in k means determains how many clusters we need to use to determain that we use elbow method


```python
max_k = 12

inertia = []

k_values = range(2,max_k+1)

silhoutte_score = []

for k in k_values:
  kmeans = KMeans(n_clusters=k , random_state=42, max_iter=1000)

  cluster_labels = kmeans.fit_predict(scaled_df)

  sil_score = silhouette_score(scaled_df,cluster_labels)

  silhoutte_score.append(sil_score)

  inertia.append(kmeans.inertia_)


plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(k_values , inertia, marker='o')
plt.xlabel("No. of Clusters (k) ")
plt.ylabel('Inertia')
plt.title("KMeans Inertia fro Different Values of K")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(k_values ,silhoutte_score , marker='o')
plt.xlabel("No. of Clusters (k) ")
plt.ylabel('Inertia')
plt.title("KMeans Inertia fro Different Values of K")
plt.grid(True)


plt.show()
```

We can use Silhouette Score to determain the accurate cluster count

# we choose 4 as the cluster

```python
kmeans = KMeans(n_clusters=4, random_state=42, max_iter=1000)

cluster_labels = kmeans.fit_predict(scaled_df)

cluster_labels
```

# Outlier Data analysis

```python
overlap_index = Monetary_outliers_df.index.intersection(frequency_outliers_df.index)

overlap_index
```

```python
monetray_only_outliers = Monetary_outliers_df.drop(overlap_index)
frequency_only_outliers = frequency_outliers_df.drop(overlap_index)
monetray_and_frequency_only_outliers = Monetary_outliers_df.loc[overlap_index]

monetray_only_outliers["Cluster"] = -1
frequency_only_outliers["Cluster"] = -2
monetray_and_frequency_only_outliers["Cluster"] = -3

outlier_clusters_df = pd.concat([monetray_only_outliers
                                 ,frequency_only_outliers
                                 ,monetray_and_frequency_only_outliers])

outlier_clusters_df
```

```python
cluster_colors= {
    -1:"purple",
    -2:"brown",
    -3:"black",
}

plt.figure(figsize=(12,18))

plt.subplot(3,1,1)
sns.violinplot(x=outlier_clusters_df["Cluster"], y=outlier_clusters_df["MonetaryValue"], palette= cluster_colors , hue= outlier_clusters_df['Cluster']  )
# sns.violinplot(y=non_outliers_df["MonetaryValue"] , color = 'grey' , linewidth=1.0 )

plt.subplot(3,1,2)
sns.violinplot(x=outlier_clusters_df["Cluster"], y=outlier_clusters_df["Frequency"], palette= cluster_colors , hue= outlier_clusters_df['Cluster'] )
# sns.violinplot(y=non_outliers_df["Frequency"] , color = 'grey' , linewidth=1.0 )

plt.subplot(3,1,3)
sns.violinplot(x=outlier_clusters_df["Cluster"], y=outlier_clusters_df["Recency"], palette= cluster_colors , hue=outlier_clusters_df['Cluster'] )
# sns.violinplot(y=non_outliers_df["Recency"] , color = 'grey' , linewidth=1.0 )


plt.show()
```

#  outlier customer type

```python
cluster_labels = {
-1 : 'rare-high-spenders',
-2 : 'frequent-low-spenders',
-3 : 'frequent-high-spenders',
0 :  'high-frequent-not-recent-spenders',
1 : 'low-infrequent-non-recent-spenders',
2 : 'low-leastactive-recent-spenders',
3 : 'high-frequent-recent-spenders'
}

cluster_labels
```

```python
full_cluster_df = pd.concat([non_outliers_df,outlier_clusters_df])

full_cluster_df["ClusterName"] = full_cluster_df["Cluster"].map(cluster_labels)

full_cluster_df
```

```python
cluster_counts = full_cluster_df['ClusterName'].value_counts()
feature_means = full_cluster_df.groupby('ClusterName')[['Recency','Frequency','MonetaryValue']].mean()

fig , ax1 = plt.subplots(figsize=(25,8))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, ax=ax1 , palette ='viridis')
ax1.set_xlabel('Number of customers', color ='b')
ax1.set_ylabel('Cluster Distribution with Avg Feature Values')

ax2 = ax1.twinx()

sns.lineplot(data=feature_means,ax=ax2, palette='Set2', marker='o')
ax2.set_ylabel('Avg Feature Values')

plt.show()
```
